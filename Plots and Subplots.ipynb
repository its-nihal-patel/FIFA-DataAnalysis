```python
# importing required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# import the data
fifa = pd.read_csv("players_20.csv")
```

```python
fifa.head()
```

```python
fifa.shape
```

```python
# print all the column labels
for col in fifa.columns:
    print(col)
```

```python
# country wise majority players
fifa['nationality'].value_counts()
```

```python
# top 10 countries with most players
fifa['nationality'].value_counts()[0:10]
```

## Plots for top 10 Countries with most players
### Line Plot

```python
x = fifa['nationality'].value_counts()[0:10]
y = x.sort_values()

# Create a line graph
plt.figure(figsize=(9, 5))
plt.plot(x.index, y.values)

# customize the plot
plt.title('Line Plot of Top 10 Countries with the Most Players')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.show()
```

### Bar Plot

```python
plt.figure(figsize=(9,5))

# plot the data
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()), 
        list(fifa['nationality'].value_counts()[0:10]), color='r')

# customize the plot
plt.title('Bar plot of Top 10 Countries with the Most Players')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.show()
```

### Histogram plot

```python
# Filter the DataFrame to include only the top 10 countries
filtered_df = fifa[fifa['nationality'].isin(x.index)]

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(filtered_df['nationality'], bins=10, color='purple')

# customize the plot
plt.title('Histogram of Top 10 Countries with the Most Players')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.show()
```

## Scatter plot

```python
plt.figure(figsize=(9, 6))

# plot the data
plt.scatter(list(fifa['nationality'].value_counts()[0:20].keys()),
            list(fifa['nationality'].value_counts()[0:20]), color='g')

# customize the plot
plt.title('Scatter plot of Top 10 Countries with the Most Players')
plt.xlabel('Nationality')
plt.ylabel('Number of players')

# rotate x-axis labels for better visibility
plt.xticks(rotation=90)
plt.show()
```

 # Subplots

```python
# create a plot with 4 empty subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(18, 10))
```

### Plotting above plots to each different axis

```python
# create a plot with 4 empty subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(18, 10))

# Line plot
ax1.plot(x.index, y.values)
ax1.set_title('Line Plot of Top 10 Countries with the Most Players')
ax1.set_xlabel('Nationality')
ax1.set_ylabel('Number of Players')

# Bar plot
ax2.bar(list(fifa['nationality'].value_counts()[0:10].keys()),
        list(fifa['nationality'].value_counts()[0:10]), color='r')
ax2.set_title('Bar plot of Top 10 Countries with the Most Players')
ax2.set_xlabel('Nationality')
ax2.set_ylabel('Number of Players')

# Histogram
top_10_countries = fifa['nationality'].value_counts().head(10)
filtered_df = fifa[fifa['nationality'].isin(top_10_countries.index)]

ax3.hist(filtered_df['nationality'], bins=10, color='purple')
ax3.set_title('Histogram of Top 10 Countries with the Most Players')
ax3.set_xlabel('Nationality')
ax3.set_ylabel('Number of Players')

# Scatter plot
ax4.scatter(list(fifa['nationality'].value_counts()[0:20].keys()),
            list(fifa['nationality'].value_counts()[0:20]), color='g')
ax4.set_title('Scatter plot of Top 10 Countries with the Most Players')
ax4.set_xlabel('Nationality')
ax4.set_ylabel('Number of Players')
ax4.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

## Top 5 HIghest Rated Players w.r.t. Shooting Skills

```python
plyr_shooting = fifa[['short_name', 'shooting']]
best_shooting_players = plyr_shooting.sort_values(by=['shooting'], ascending=False).head()
best_shooting_players
```

```python
plt.figure(figsize=(10, 6))

# plot the data
plt.bar(best_shooting_players['short_name'], best_shooting_players['shooting'], color='black')

# customize the plot
plt.title('Top 5 HIghest Rated Players w.r.t. Shooting Skills')
plt.xlabel('Player')
plt.ylabel('Shooting Skill')

# rotate x-axis labels for better visibility
plt.xticks(rotation=90)
plt.show()
```

### The difference in shooting skills is minor among best playes, It is difficult to visualize it effectively using bar plot.
- So plotting horizontal bar plot to highlight the subtle differences

```python
plt.figure(figsize=(10, 6))

# plot the data
plt.barh(best_shooting_players['short_name'], best_shooting_players['shooting'], color='b')

# customize the plot
plt.title('Best-rated Players for Shooting Skills')
plt.xlabel('Shooting Skill')
plt.ylabel('Player')
plt.show()
```

### Still Not Effective Visualization. So, Using Line Graph.

```python
plt.figure(figsize=(9, 6))

# plot data
plt.plot(best_shooting_players['short_name'], best_shooting_players['shooting'], marker='o')

# customize the plot
plt.title('Top 5 HIghest Rated Players w.r.t. Shooting Skills')
plt.xlabel('Player')
plt.ylabel('Shooting Skill')

# rotate x-axis labels for better visibility
plt.xticks(rotation=90)
plt.show()
```

## Top 5 HIghest Rated Players w.r.t. Defending Skills

```python
player_defending = fifa[['short_name', 'defending', 'club']]
player_defending.sort_values(by=['defending'], ascending=False).head()
```

```python
# Sort the players by defending rating in descending order
sorted_players = player_defending.sort_values(by='defending', ascending=False).head()

plt.figure(figsize=(10, 6))

# plot the data
sns.barplot(x='short_name', y='defending', hue='club', data=sorted_players)

# customize the plot
plt.title('Highest-Rated Players by Defending Skills')
plt.xlabel('Player')
plt.ylabel('Defending Rating')

# Set the y-axis range
plt.ylim(50, 100)
plt.xticks(rotation=90)

# Adjust legend style and position
plt.legend(title='Club', loc='upper right', ncol=len(sorted_players['club'].unique()))
plt.show()
```
